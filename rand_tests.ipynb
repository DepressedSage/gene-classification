```python
import numpy as np

# List of protein sequences
sequences = [
    "AARCIDGAFRPKATDG",
    "ACDPHSGHFV",
    "ADDVKLWRQRGMSTW",
    "ADKPDESTL",
    "AEENDEENVEEVEENVEENV",
    "AEHQAIIRDVLTASD"
]

# Create a dictionary to map each amino acid to an index
amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
aa_to_index = {aa: i for i, aa in enumerate(amino_acids)}

# Function to convert a sequence to a one-hot encoding
def sequence_to_one_hot(sequence):
    one_hot_encoding = np.zeros((len(sequence), len(amino_acids)))
    for i, aa in enumerate(sequence):
        if aa in aa_to_index:
            one_hot_encoding[i, aa_to_index[aa]] = 1
    return one_hot_encoding

# Create one-hot encodings for all sequences
one_hot_encodings = [sequence_to_one_hot(seq) for seq in sequences]

# Print one-hot encodings for each sequence
for i, seq in enumerate(sequences):
    print(f"Sequence {i + 1}:")
    print(one_hot_encodings[i])
    print()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

# List of protein sequences
sequences = [
    "AARCIDGAFRPKATDG",
    "ACDPHSGHFV",
    "ADDVKLWRQRGMSTW",
    "ADKPDESTL",
    "AEENDEENVEEVEENVEENV",
    "AEHQAIIRDVLTASD"
]

# Join all sequences into a single string
all_amino_acids = ''.join(sequences)

# Get unique amino acids
unique_amino_acids = sorted(set(all_amino_acids))

# Reshape sequences into a 2D array with one row per sequence
reshaped_sequences = [[aa] for aa in all_amino_acids]

# Create one-hot encoder
one_hot_encoder = OneHotEncoder(categories=[unique_amino_acids], sparse=False)

# Fit and transform on reshaped sequences
one_hot_encodings = one_hot_encoder.fit_transform(reshaped_sequences)

# Print one-hot encodings for each sequence
for i, seq in enumerate(sequences):
    print(f"Sequence {i + 1}:")
    print(one_hot_encodings[i])
    print()
```

```python
import torch
from torch.nn.utils.rnn import pad_sequence

# List of protein sequences
sequences = [
    "AARCIDGAFRPKATDG",
    "ACDPHSGHFV",
    "ADDVKLWRQRGMSTW",
    "ADKPDESTL",
    "AEENDEENVEEVEENVEENV",
    "AEHQAIIRDVLTASD"
]

# Define the set of unique characters (amino acids)
unique_characters = sorted(set(''.join(sequences)))

# Additional tokens
pad_token = '<PAD>'
start_token = '<START>'
end_token = '<END>'
unknown_token = '<UNK>'

# Create a mapping from characters to indices
char_to_index = {char: i + 4 for i, char in enumerate(unique_characters)}
char_to_index[pad_token] = 0
char_to_index[start_token] = 1
char_to_index[end_token] = 2
char_to_index[unknown_token] = 3

# Convert each sequence to its index representation
indexed_sequences = [[char_to_index.get(char, char_to_index[unknown_token]) for char in seq] for seq in sequences]

# Convert indexed sequences to PyTorch tensors
tensor_sequences = [torch.tensor(seq) for seq in indexed_sequences]

# Pad sequences to a fixed length
padded_sequences = pad_sequence(tensor_sequences, batch_first=True, padding_value=char_to_index[pad_token])

# Print padded sequences
print(padded_sequences)
```
